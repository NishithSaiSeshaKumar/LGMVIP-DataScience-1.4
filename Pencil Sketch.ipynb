# Pencil Sketch Generation

Let's create a pencil sketch from a given image using image processing techniques.

```python
# Import necessary libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Step 1: Read and Preprocess the Image
Let's start by reading the image and converting it to the RGB color space.

```python
# Read the image
image_path = 'dog4.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## Step 2: Convert to Grayscale
Convert the RGB image to grayscale, which is a single-channel image.

```python
# Convert to grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

## Step 3: Invert Grayscale Image
Create a negative or inverted image to enhance details.

```python
# Invert grayscale image
inverted_image = cv2.bitwise_not(grayscale_image)
```

## Step 4: Apply Gaussian Blur
Apply Gaussian blur to the inverted image to prepare for mixing.

```python
# Apply Gaussian blur to inverted image
blurred_image = cv2.GaussianBlur(inverted_image, (51, 51), 0)
```

## Step 5: Invert Blurred Image
Invert the blurred image to get the final inverted blurred image.

```python
# Invert the blurred image
inverted_blurred_image = cv2.bitwise_not(blurred_image)
```

## Step 6: Create Pencil Sketch
Mix the grayscale image with the inverted blurred image to create the pencil sketch.

```python
# Create pencil sketch
pencil_sketch_image = cv2.divide(grayscale_image, inverted_blurred_image, scale=256.0)
```

## Step 7: Display Results
Display the original image and the generated pencil sketch side by side.

```python
# Display images
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(pencil_sketch_image, cmap='gray')
plt.title('Pencil Sketch')

# Remove axis labels
plt.axis('off')

# Show the plot
plt.show()
```
